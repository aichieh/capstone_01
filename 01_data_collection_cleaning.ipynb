### Download files from National Health and Nutrition Examination Survey (NHANES)

```python
# credit: https://www.w3schools.com/python/ref_requests_response.asp
# credit: https://github.com/psf/requests/issues/6346
# credit: https://scrapfly.io/blog/web-scraping-with-python-beautifulsoup/
# credit: https://stackoverflow.com/questions/16694907/download-large-file-in-python-with-requests

# Create an output folder.
os.mkdir("output")

def download_file(url, dest_folder):  
    if not os.path.exists(dest_folder):
        os.makedirs(dest_folder)  # create folder if it does not exist

       
    filename = url.split('/')[-1].replace(" ", "_") 
    file_path = os.path.join(dest_folder, filename)

    res = requests.get(url, stream=True)    # set up response  
    if res.ok:
        print("saving to", os.path.abspath(file_path))
        with open(file_path, 'wb') as f:
            for chunk in res.iter_content(chunk_size=1024 * 8):
                if chunk:
                    f.write(chunk)
                    f.flush()
                    os.fsync(f.fileno())
    else:               # check for errors and raise ConnectionError if there is one
        raise ConnectionError(f'Error code:{res.status_code} has occured.')
```

```python
def get_data(url, component):
    res = requests.get(url)
    soup = BeautifulSoup(res.text, "html.parser")
    for i in soup.body.table.find_all("a",href=True):
        s = str(i['href'])
        if "XPT" in s:
            download_file(url+s, component)
    df = pd.read_html(str(soup.body.table))[0]
    df.to_csv("./output/full_"+component+".csv")
```
